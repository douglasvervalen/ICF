```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rc('lines', 
       linewidth=3, 
       linestyle='-'
       )
plt.rc('axes', 
       linewidth=1.5, 
       labelsize=15
       )
plt.rc('xtick', 
       direction='in', 
       labelsize=15
       )
plt.rc('ytick', 
       direction='in', 
       labelsize=15,
       right='True'
       )
plt.rc('legend', 
       frameon='False', 
       fontsize=15,
       loc='best'
       )
```

```python
def f(x,a):
  y = x
  return y

def g(x,a):
  y = a*x**2
  return y

def h(x):
    return np.e**x
def j(x):
    return np.sin(x)
def k(x):
    return np.cos(x)
```

```python
fig, ax = plt.subplots(2,2,figsize=(18,18))

a=0.2
b=1.2

x0 = np.linspace(-2,2,40)
x1 = np.linspace(-6.5,6.5,40)

ax[0][0].plot(x0,f(x0,a),label='f(x, 10, 0)')
ax[0][0].set_xlabel('eixo-x1')
ax[0][0].set_xticks([-2,-1,0,1,2])
ax[0][0].set_yticks([-2,-1,0,1,2])
ax[0][0].set_ylabel('eixo-y1')
ax[0][0].legend()

ax[0][1].plot(x0,g(x0,b),label='g(x,1,0,0)')
ax[0][1].set_xlabel('eixo-x2')
ax[0][1].set_ylabel('eixo-y2')
ax[0][1].set_xticks([-2,-1,0,1,2])
ax[0][1].set_yticks([0,1,2,3,4])
ax[0][1].legend()

ax[1][0].plot(x0,h(x0),label='exp(x)')
ax[1][0].plot(x0,h(-x0),label='exp(-x)', linestyle='--')
ax[1][0].set_xlabel('eixo-x3')
ax[1][0].set_ylabel('eixo-y3')
ax[1][0].legend()

ax[1][1].plot(x1,j(x1),label='sin(x)')
ax[1][1].plot(x1,k(x1),label='cos(x)',linestyle='--')
ax[1][1].set_xlabel('eixo-x4')
ax[1][1].set_xticks([-10,-5,0,5,10])
ax[1][1].set_yticks([-1,-0.5,0,0.5,1])
ax[1][1].set_ylabel('eixo-y4')
ax[1][1].legend()

plt.show()
```
