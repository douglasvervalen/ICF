```python
import matplotlib.pyplot as plt
import numpy as np 
from scipy.optimize import curve_fit
from numpy.random import default_rng
```

```python
def f1(x):
    return np.exp(-x)

def f2(x):
    return np.exp(-2*x)


def f3(x):
    return np.exp(-3*x)

def func_linear(x,a,b):
    return a*x+b

def fe(x,a,b):
    return np.exp(a*x)+b
```

```python
xp = np.linspace(0,4,65)
fig, ax = plt.subplots(1,2, figsize=(10,7))
ax[0].plot(xp, f1(xp), color="blue",label='e^(-x)', linewidth=4)
ax[0].plot(xp, f2(xp), color="orange",label='e^(-2x)',linewidth=4)
ax[0].plot(xp, f3(xp), color="green",label='e^(-3x)',linewidth=4)
ax[0].set_xticks([0,2,4])
ax[0].set_ylabel("Decaimento exponencial")
ax[0].set_xlabel("Tempo")
ax[0].legend()

ax[1].plot(xp, f1(xp), color="blue", linewidth=4)
ax[1].plot(xp, f2(xp), color="orange", linewidth=4)
ax[1].plot(xp, f3(xp), color="green", linewidth=4)
ax[1].set_yscale("log")
ax[1].set_ylim(0.000001,1)
ax[1].set_xticks([0,2,4])
ax[1].set_xlabel("Tempo")
plt.legend()
plt.show()
```

```python
xp = np.linspace(0,4,65)
rng = default_rng(12345)
x_noise = 0.1*rng.standard_normal(size=xp.size)
y = xp - x_noise
fig, ax = plt.subplots(2,2, figsize=(14,14))

ax[0][0].plot(xp, f1(xp), color="blue", label='e^(-x)', linewidth=4)
ax[0][0].plot(xp, f2(xp), color="orange",label='e^(-2x)',linewidth=4)
ax[0][0].plot(xp, f3(xp), color="green",label='e^(-3x)',linewidth=4)
ax[0][0].set_xticks([0,2,4])
ax[0][0].set_ylabel("Decaimento exponencial")
ax[0][0].set_xlabel("Tempo")
ax[0][0].legend()

ax[0][1].plot(xp, f1(xp), color="blue",linewidth=4)
ax[0][1].plot(xp, f2(xp), color="orange",linewidth=4)
ax[0][1].plot(xp, f3(xp), color="green",linewidth=4)
ax[0][1].set_yscale("log")
ax[0][1].set_ylim(0.0000001,10)
ax[0][1].set_xticks([0,2,4])
ax[0][1].set_xlabel("Tempo")

ax[1][0].plot(xp, f1(y), color="blue",label='e^(-x + erro)',linewidth=4)
ax[1][0].plot(xp, f2(y), color="orange",label='e^(-2x + erro)',linewidth=4)
ax[1][0].plot(xp, f3(y), color="green",label='e^(-3x + erro)',linewidth=4)
ax[1][0].set_xticks([0,2,4])
ax[1][0].set_ylabel("Decaimento exponencial")
ax[1][0].set_xlabel("Tempo")
ax[1][0].legend()

ax[1][1].plot(xp, f1(y), color="blue",linewidth=4)
ax[1][1].plot(xp, f2(y), color="orange",linewidth=4)
ax[1][1].plot(xp, f3(y), color="green",linewidth=4)
ax[1][1].set_yscale("log")
ax[1][1].set_ylim(0.0000001,10)
ax[1][1].set_xticks([0,2,4])
ax[1][1].set_xlabel("Tempo")
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots(1,2, figsize=(14,14))
popt1, pcov = curve_fit(fe,xp,f1(y))
popt2, pcov = curve_fit(fe,xp,f2(y))
popt3, pcov = curve_fit(fe,xp,f3(y))

ax[0].plot(xp, f1(y), color="blue",label='e^(-x + erro)',linewidth=4)
ax[0].plot(xp, f2(y), color="orange",label='e^(-2x + erro)',linewidth=4)
ax[0].plot(xp, f3(y), color="green",label='e^(-3x + erro)',linewidth=4)
ax[0].plot(xp, fe(xp, *popt1), color="black",label='e^(-x + erro)',linestyle='--',linewidth=4)
ax[0].plot(xp, fe(xp, *popt2), color="black",label='e^(-2x + erro)',linestyle='--',linewidth=4)
#ax[0].plot(xp, fe(xp, *popt3), color="black",label='e^(-3x + erro)',linestyle='--',linewidth=4)
ax[0].set_xticks([0,2,4])
ax[0].set_ylabel("Decaimento exponencial")
ax[0].set_xlabel("Tempo")
ax[0].legend()

ax[1].plot(xp, f1(y), color="blue",linewidth=4)
ax[1].plot(xp, f2(y), color="orange",linewidth=4)
ax[1].plot(xp, f3(y), color="green",linewidth=4)
ax[1].plot(xp, fe(xp, *popt1), color="black",label='e^(-x + erro)',linestyle='--',linewidth=4)
#ax[1].plot(xp, fe(xp, *popt2), color="black",label='e^(-2x + erro)',linestyle='--',linewidth=4)
#ax[1].plot(xp, fe(xp, *popt3), color="black",label='e^(-3x + erro)',linestyle='--',linewidth=4)
ax[1].set_yscale("log")
ax[1].set_ylim(0.0000001,10)
ax[1].set_xticks([0,2,4])
ax[1].set_xlabel("Tempo")
plt.legend()
plt.show()
```
