# <h3 align="center" id="heading">Alcance máximo com  h>0</h3>


A equação $y(x)$ é dada por: $$ y(x)=\tan(\alpha)x-\left(\frac{g}{2v_0^2\cos^2(\alpha)}\right)x^2$$

Isolando o x temos:
$$
x = \frac{2\cdot v_0^2\cdot \cos^2(\theta)}{g} \cdot \left(- \tan(\theta) \pm \sqrt{\tan(\theta)^2 - \frac{4\cdot h\cdot g}{2 \cdot v_0^2 \cdot \cos^2(\theta)}} \right)
$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f_R2(v0,alpha,y0):
  g = 9.81 
  alpha_rad = alpha * ( np.pi/180) 
  y = ((v0*np.cos(alpha_rad))/g)*(v0*np.sin(alpha_rad)+np.sqrt((v0**2)*np.sin(alpha_rad)**2+2*g*y0))
  return y
```

```python
theta = np.linspace(0,90,30)
xtick = [30, 35, 40, 45, 50, 55, 60]
plt.figure(figsize=(8,5))
plt.plot(theta,f_R2(1,theta,0),marker='o')
plt.plot(theta,f_R2(1,theta,0.005),marker='o')
plt.plot(theta,f_R2(1,theta,0.01),marker='o')
plt.ylabel("Alcance [m]")
plt.xlabel("ângulo [graus]")
plt.grid()
plt.show()
```

## Procurando pelos valores dos angulos que maximizam os alcances para os casos: h=0, h=0.01 e h=0.005; Temos:

$$v_0 = cte$$

```python
j = 41.0
valores = []
angulo = []
for i in range(0, 5000):
    valores.append(f_R2(1,j,0))
    angulo.append(j)
    j += 0.001
maxval = valores.index(max(valores))
print(f'Angulo: {angulo[maxval]:.3f}')
```
