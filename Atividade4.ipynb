```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import get_test_data
```

```python
fig = plt.figure(figsize=(14,8))
ax1 = fig.add_subplot(1,2,1,projection='3d')
x = np.arange(-3, 3, 0.05)
y = np.arange(-3, 3, 0.05)
X, Y = np.meshgrid(x, y)
Z1 = np.exp(-(X**2 + Y**2) / 2) / (2 * np.pi)
ax1.plot_wireframe(X,Y,Z1,rcount=10000,ccount=10000,label='wireframe1',color='orange',linewidth=0.7)
ax1.set_xlabel('eixo x')
ax1.set_ylabel('eixo y')
ax1.set_zlabel('eixo z')
ax1.set_xlim(-3,3)
ax1.set_ylim(-3,3)
ax1.set_zlim(-0.1,0.1)
ax1.legend()

ax2 = fig.add_subplot(1,2,2,projection='3d')
Z2 = (np.exp(-(((X - 1) / 1.5)**2 + ((Y - 1) / 0.5)**2) / 2) /(2 * np.pi * 0.5 * 1.5))
Z = Z2 - Z1
ax2.plot_wireframe(X,Y,Z,rcount=10000,ccount=10000,label='wireframe2',color='black', linewidth=0.5)
ax2.set_xlabel('eixo x')
ax2.set_ylabel('eixo y')
ax2.set_zlabel('eixo z')
ax2.set_xlim(-3,3)
ax2.set_ylim(-3,3)
ax2.set_zlim(-0.1,0.1)
ax2.legend()

plt.show()
```
